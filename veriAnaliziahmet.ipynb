```python
import re
```

```python
def unvanCikar (veri): # regex duzenli ifade regexr.com ayrıntılı bilgi
  veri=re.sub(".*,.([^.]*)\\..*","\\1",veri) # virgül görene kadar git, virgülden sonra bir kar
  return veri
```

```python
unvanCikar ("Bilgisayar Muh, Kıdemli.")
```

```python
unvan = df["Meslek"].apply(unvanCikar)
unvan
```

```python
df["Brans"] = unvan
df
```
